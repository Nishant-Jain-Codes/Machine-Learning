4.  What is the difference between `range(10)`, `range(0, 10)`, and `range(0, 10, 1)` in a for loop?

answer: They all do the same thing. The range(10) call ranges from 0 up to (but not including) 10, range(0, 10) explicitly tells the loop to start at 0, and range(0, 10, 1) explicitly tells the loop to increase the variable by 1 on each iteration.

````python
#p1
import random
tries = int(input('How many tries do you want? '))
no = random.randint(1, 20)
for i in range(tries):
    guess = int(input('Guess a number between 1 and 20: '))
    if guess == no:
        print('You guessed it!')
        break
    elif guess > no:
        print('Too high!')
    else:
        print('Too low!')
else:
    print('You did not guess it! the number was', no)

````

````python
#p2
import random
total_plays = input('How many times do you want to play? ')
player_score = 0
computer_score = 0
for i in range(int(total_plays)):
    player = input('Rock, Paper or Scissors? ')
    computer = random.choice(['Rock', 'Paper', 'Scissors'])
    print('You chose', player, 'and the computer chose', computer)
    if player == computer:
        print('Draw!')
    elif player == 'Rock' and computer == 'Scissors':
        print('You win!')
        player_score += 1
    elif player == 'Paper' and computer == 'Rock':
        print('You win!')
        player_score += 1
    elif player == 'Scissors' and computer == 'Paper':
        print('You win!')
        player_score += 1
    else:
        print('You lose!')
        computer_score += 1
print('The final score is', player_score, 'to', computer_score)
````

````python
import random
word = random.choice(['chicken', 'dog', 'cat', 'mouse', 'frog'])
guessed = []
tries = len('Hangman')
while tries > 0:
    out = ''
    for letter in word:
        if letter in guessed:
            out = out + letter
        else:
            out = out + '_'
    if out == word:
        break
    print('Guess the word:', out)
    print(tries, 'chances left')
    guess = input()
    guessed.append(guess)
    if guess not in word:
        tries = tries - 1
        print('Wrong! You lose a chance.')
        print()
````
