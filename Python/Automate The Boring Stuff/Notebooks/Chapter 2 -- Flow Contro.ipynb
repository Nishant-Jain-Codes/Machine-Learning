3. Write out the truth tables of each Boolean operator (that is, every possible combination of Boolean values for the operator and what they evaluate to).

answer : 
```python
True and True = True
True and False = False
False and True = False
False and False = False

True or True = True
True or False = True
False or True = True
False or False = False

not True = False
not False = True
```

4.  What is the difference between `range(10)`, `range(0, 10)`, and `range(0, 10, 1)` in a for loop?

answer: They all do the same thing. The range(10) call ranges from 0 up to (but not including) 10, range(0, 10) explicitly tells the loop to start at 0, and range(0, 10, 1) explicitly tells the loop to increase the variable by 1 on each iteration.

5.  Write a short program that prints the numbers 1 to 10 using a for loop. Then write an equivalent program that prints the numbers 1 to 10 using a while loop.

answer:
```python
for i in range(1,11):
    print(i)
```
```python
i = 1
while i < 11:
    print(i)
    i += 1
```
